#  1. Univariate Analysis:
* Univariate analysis refers to the analysis of a single variable. The goal is to understand the distribution, spread, and central tendency of that variable.

# Histogram

```python
sns.set(style='darkgrid')
plt.figure(figsize=(10,8))
plt.hist(df['review_scores_rating'])
ax=plt.gca()
for bar in ax.patches:
    x=bar.get_x()+bar.get_width()/2
    y=bar.get_height()
    ax.text(x,y,f'{y:.0f}',ha='center',va='bottom',fontsize=10)
plt.title('Review_Scores_Rating',fontsize=16)
plt.xlabel('Ratings',fontsize=12)
plt.ylabel('Count of ratings',fontsize=12)
plt.savefig('Review_Scores.png',format='png')
plt.show()
```

# CountPlot

```python
sns.set(style='darkgrid')
plt.figure(figsize=(8,6))
color=['red','blue','green']
sns.countplot(data=df,x='room_type',hue='room_type',palette=color)
ax=plt.gca()
for bar in ax.patches:
    x=bar.get_x()+bar.get_width()/2
    y=bar.get_height()
    ax.text(x,y,f'{y:.0f}',ha='center',va='bottom',fontsize=10)

#Adding title and labels
plt.title('Count of Rooms',fontsize=16)
plt.xlabel('Room types',fontsize=14)
plt.ylabel('Count of Rooms',fontsize=12)
plt.savefig('countplot.png',format='png')
plt.show()
```

# Total  Bathrooms

# Bar Plot

```python
sns.set(style='darkgrid')
grouped=df.groupby('room_type')['bathrooms'].sum().reset_index()
plt.figure(figsize=(10,8))
color=['yellow','red','black']
sns.barplot(data=grouped,x='bathrooms',y='room_type',hue='room_type',palette=color,orient='h')
ax=plt.gca() #Get Current axes
for bar in ax.patches:
    x = bar.get_width()  # For horizontal, we use width for x position
    y = bar.get_y() + bar.get_height() / 2  # For horizontal, we use y position
    ax.text(x, y, f'{x:.0f}', ha='left', va='center', fontsize=14)

plt.title("Count Of Bathrooms For Each room types",fontsize=16)
plt.xlabel('Count of Bathrooms',fontsize=14)
plt.ylabel('Room Types',fontsize=14)
plt.savefig('barplot.png',format='png')
plt.show()
```

## Pie Chart

```python
room_type = df.groupby('room_type')['review_scores_rating'].mean()
#Create Piechart
plt.figure(figsize=(8,8))
plt.pie(room_type,labels=room_type.index,
       autopct='%1.1f%%',startangle=120,
       colors=sns.color_palette('pastel'))
plt.title('review_scores_rating for Rooms',fontsize=16)
plt.savefig('Review Scores Rating for Rooms.png',format='png')
plt.show()
```

# Heat Map

```python
numeric_data = df[['accommodates','bathrooms','review_scores_rating','bedrooms','beds','log_price']]

corr_matrix = numeric_data.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',
           fmt='.2f',linewidth=0.5)

plt.title('Correlation of heat map of Numeric Variables',fontsize=14)
plt.savefig('Heat map.png',format='png')
plt.show()
```
